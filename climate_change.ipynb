# 1.0 About Dataset
## Description:

This dataset provides a comprehensive overview of key climate change indicators collected across different countries from the year 2000 to 2024. It includes 1000 data points capturing various environmental and socio-economic factors that reflect the global impact of climate change. The dataset focuses on average temperature, CO2 emissions, sea-level rise, rainfall patterns, and more, enabling users to analyze trends, correlations, and anomalies.

## Fields Explanation:

### Year: 
The year in which the data was recorded, ranging from 2000 to 2024. It helps track historical trends in climate change and related variables over time.
Country: The country or region where the climate data was collected. The dataset includes a diverse set of countries from across the globe, representing different geographic regions and climates.
### Average Temperature (°C): 
The average annual temperature recorded in each country, measured in degrees Celsius. This field allows for comparisons of temperature changes across regions and time.
### CO2 Emissions (Metric Tons per Capita): 
The average amount of CO2 emissions per capita in metric tons, reflecting the country's contribution to greenhouse gases. This field is useful for analyzing the link between human activity and environmental changes.
### Sea Level Rise (mm): 
The recorded annual sea-level rise in millimeters for coastal regions. This indicator reflects the global warming effect on melting glaciers and thermal expansion of seawater, critical for studying impacts on coastal populations.
### Rainfall (mm): 
The total annual rainfall recorded in millimeters. This field highlights changing precipitation patterns, essential for understanding droughts, floods, and water resource management.
### Population: 
The population of the country in the given year. Population data is important to normalize emissions or other per-capita analyses and understand human impact on the environment.
### Renewable Energy (%): 
The percentage of total energy consumption in a country that comes from renewable energy sources (solar, wind, hydro, etc.). This metric is vital for assessing the progress made toward sustainable energy and reducing reliance on fossil fuels.
### Extreme Weather Events: 
The number of extreme weather events recorded in each country, such as hurricanes, floods, wildfires, and droughts. Tracking these events helps correlate the increase in climate change with the frequency of natural disasters.
### Forest Area (%): 
The percentage of the total land area of a country covered by forests. Forest cover is a critical indicator of biodiversity and carbon sequestration, with reductions often linked to deforestation and habitat loss.

# Detecting outlier

```python
# Numerical columns
num_cols = [
    "avg_temperature",
    "co2_emissions",
    "sea_level_rise",
    "rainfall",
    "population",
    "renewable_energy",
    "forest_area",
    "extreme_weather_events",
]

# -------- Outlier Detection (IQR) --------
print("=== OUTLIER DETECTION USING IQR METHOD ===\n")

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}")
    print(f"  Q1 = {Q1:.2f}")
    print(f"  Q3 = {Q3:.2f}")
    print(f"  IQR = {IQR:.2f}")
    print(f"  Lower Bound = {lower_bound:.2f}")
    print(f"  Upper Bound = {upper_bound:.2f}")
    print(f"  Number of Outliers = {len(outliers)}")
    print("-" * 50)
```

# co2 emissions and avarage temperature

```python
import matplotlib.pyplot as plt
import numpy as np

# ===============================
# Make sure Region is assigned
# ===============================
def assign_region(country):
    for region, countries in regions.items():
        if country in countries:
            return region
    return "Other"

df["Region"] = df["country"].apply(assign_region)

df_region = df[df["Region"].isin(
    ["Asia", "Europe", "Africa", "Americas", "Oceania"]
)]

# ===============================
# Aggregate averages by continent
# ===============================
region_avg = (
    df_region
    .groupby("Region")
    .agg({
        "co2_emissions": "mean",
        "avg_temperature": "mean"
    })
    .reset_index()
)

# ===============================
# Bar positions
# ===============================
x = np.arange(len(region_avg["Region"]))
width = 0.35

# ===============================
# Plot
# ===============================
plt.figure(figsize=(10, 6))

plt.bar(
    x - width / 2,
    region_avg["co2_emissions"],
    width,
    label="CO2 Emissions",
    color="steelblue"
)

plt.bar(
    x + width / 2,
    region_avg["avg_temperature"],
    width,
    label="Average Temperature",
    color="salmon",
)

# ===============================
# Labels & layout
# ===============================
plt.xlabel("Continent")
plt.ylabel("Average Value")
plt.title("CO2 Emissions vs Average Temperature by Continent")
plt.xticks(x, region_avg["Region"])
plt.legend()
plt.grid(True, axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ===============================
# Prepare yearly averages
# ===============================
yearly_avg = (
    df.groupby("year")
    .agg({
        "co2_emissions": "mean",
        "avg_temperature": "mean"
    })
    .reset_index()
)

# Optional: keep every 5 years for clarity
yearly_avg = yearly_avg[yearly_avg["year"] % 5 == 0]

# Reshape data for seaborn
plot_data = yearly_avg.melt(
    id_vars="year",
    value_vars=["co2_emissions", "avg_temperature"],
    var_name="Variable",
    value_name="Value"
)

# ===============================
# Bar plot
# ===============================
plt.figure(figsize=(12, 6))

sns.barplot(
    data=plot_data,
    x="year",
    y="Value",
    hue="Variable",
    errorbar=None
)

plt.xlabel("Year")
plt.ylabel("Average Value")
plt.title("CO2 Emissions vs. Average Temperature Over Time")
plt.legend(title="Variable", labels=["CO2 Emissions", "Average Temperature"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

📊 Interpretation: Renewable Energy vs CO₂ Emissions by Country
🔹 Overall relationship

The dashed regression line slopes slightly upward, indicating a weak positive relationship between renewable energy usage and CO₂ emissions.

This suggests that, in this dataset, higher renewable energy percentages do not necessarily correspond to lower CO₂ emissions at the country level.

🔹 What this means (important)

Countries with higher renewable energy shares can still have moderate to high CO₂ emissions.

This implies that renewable energy adoption alone is not sufficient to reduce emissions if:

Total energy demand is high

Fossil fuels are still widely used

Industrial activity is strong

🌍 Country-level observations

Some countries with mid-to-high renewable energy usage still show relatively high CO₂ emissions, indicating:

Heavy industry

Large energy consumption per capita

Countries with lower renewable energy usage sometimes exhibit lower CO₂ emissions, possibly due to:

Lower industrialization

Lower energy demand

📈 Confidence band (grey area)

The wide shaded confidence interval indicates high variability among countries.

This means the relationship is not strong or consistent across all countries.

```python

# Group by Region (mean values)
grouped = (
    df.groupby("Region")[[
        "avg_temperature",
        "sea_level_rise",
        "rainfall"
    ]]
    .mean()
    .reset_index()
)

# -------- Normalize values (min-max) --------
grouped["Temp_Norm"] = (
    grouped["avg_temperature"] - grouped["avg_temperature"].min()
) / (grouped["avg_temperature"].max() - grouped["avg_temperature"].min())

grouped["SeaLevel_Norm"] = (
    grouped["sea_level_rise"] - grouped["sea_level_rise"].min()
) / (grouped["sea_level_rise"].max() - grouped["sea_level_rise"].min())

grouped["Rainfall_Norm"] = (
    grouped["rainfall"] - grouped["rainfall"].min()
) / (grouped["rainfall"].max() - grouped["rainfall"].min())
# # Bar positions
x = np.arange(len(grouped["Region"]))
width = 0.35

# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# -------- Plot 1: Temperature vs Sea Level --------
axes[0].bar(
    x - width/2,
    grouped["Temp_Norm"],
    width,
    label="Temperature",
    color="tomato"
)
axes[0].bar(
    x + width/2,
    grouped["SeaLevel_Norm"],
    width,
    label="Sea Level Rise",
    color="steelblue"
)

axes[0].set_title("Temperature vs Sea Level Rise by Region")
axes[0].set_xlabel("Region")
axes[0].set_ylabel("Normalized Average Value")
axes[0].set_xticks(x)
axes[0].set_xticklabels(grouped["Region"], rotation=45)
axes[0].legend()

# -------- Plot 2: Temperature vs Rainfall --------
axes[1].bar(
    x - width/2,
    grouped["Temp_Norm"],
    width,
    label="Temperature",
    color="tomato"
)
axes[1].bar(
    x + width/2,
    grouped["Rainfall_Norm"],
    width,
    label="Rainfall",
    color="darkgreen"
)

axes[1].set_title("Temperature vs Rainfall by Region")
axes[1].set_xlabel("Region")
axes[1].set_xticks(x)
axes[1].set_xticklabels(grouped["Region"], rotation=45)
axes[1].legend()

# Final layout
fig.suptitle("Comparison of Temperature with Sea Level Rise and Rainfall by Region")
plt.tight_layout()
plt.show()
```

```python
# Prepare country-specific DataFrames used later in the notebook
# Ensure `year` is numeric and data is sorted by year
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df = df.dropna(subset=["year"]).sort_values("year")

selected_countries = ["China", "USA", "UK"]
missing = [c for c in selected_countries if c not in df["country"].unique()]
if missing:
    print("Warning: these countries not found in dataset:", missing)

# Create copies to avoid warnings when modifying later
df_china = df[df["country"] == "China"].copy()
df_usa = df[df["country"] == "USA"].copy()
df_uk = df[df["country"] == "UK"].copy()

print(
    f"Prepared df_china: {len(df_china)} rows, df_usa: {len(df_usa)} rows, df_uk: {len(df_uk)} rows"
)
```

### 3.3 Let inspect China, USA and UK CO2 Emissions Data

```python
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))

sns.lineplot(
    data=df_china, x="year", y="co2_emissions", label="China", marker="o", linewidth=2
)

sns.lineplot(
    data=df_usa, x="year", y="co2_emissions", label="USA", marker="s", linewidth=2
)

sns.lineplot(
    data=df_uk, x="year", y="co2_emissions", label="UK", marker="^", linewidth=2
)

plt.title("CO2 Emissions Per Capita Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("CO₂ Emissions (tons per capita)", fontsize=12)

plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

### Figure14: CO₂ Emissions Per Capita Over Time

This figure illustrates the trends in CO₂ emissions per capita for China, the USA, and the UK over the observed time period. The line plot enables an easy comparison of how emissions evolve across countries and years, while the shaded areas highlight variability around each trend.

China shows the largest fluctuations in CO₂ emissions per capita, with sharp increases and decreases across several years. This indicates periods of rapid industrial growth followed by potential policy adjustments or economic slowdowns affecting emission levels.

The USA demonstrates a moderate but relatively stable trend, with emissions remaining consistently high compared to the other countries. Although fluctuations exist, the overall pattern suggests gradual changes rather than extreme volatility.

The UK exhibits a more controlled and smoother trend, with emissions generally lower than those of China and the USA. This pattern may reflect stronger environmental regulations, increased adoption of renewable energy, and long-term emission reduction strategies.

Overall, this visualization clearly shows how economic structure, energy policy, and industrial activity influence per-capita CO₂ emissions. The line plot is effective for identifying long-term trends and comparing countries, making it suitable for temporal and cross-country climate analysis.

### 3.5 Let inspect China, USA and UK Rainfall Data

### Figure16

China generally records higher rainfall values compared to the USA and the UK in many years, although several sharp declines are visible, suggesting periods of drought or reduced precipitation. The USA shows moderate rainfall levels with significant volatility, including a pronounced drop in the mid-2000s followed by periods of recovery. In contrast, the UK tends to experience lower average rainfall than China but shows several sharp increases, reflecting episodic heavy rainfall events.

The shaded areas around each line indicate the range of variability, highlighting uncertainty and dispersion in rainfall data across years. Wider shaded regions suggest greater variability, particularly for China and the USA, while the UK displays comparatively narrower ranges in some periods.

Overall, this visualization demonstrates that rainfall patterns across the three countries are inconsistent and increasingly unpredictable, which may have important implications for water resource management, agriculture, and climate resilience planning.
